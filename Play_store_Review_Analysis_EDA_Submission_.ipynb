````python
#finding out which one is better using different visualization tools
fig, ax = plt.subplots(2,1, figsize=(12,7))
sns.distplot(playstore_df['Rating'],color='firebrick',ax=ax[0])
sns.boxplot(x='Rating',data=playstore_df, ax=ax[1])
````

From the distplot visualizations, it is clear that the ratings are left skewed.

We know that if the variable is skewed, the mean is biased by the values at the far end of the distribution. Therefore, the median is a better representation of the majority of the values in the variable.

````python
playstore_df['Rating'].fillna(value=med_rating,inplace=True)
````

### 3.Removing the special characters 

Removing special characters like  '+', ',' ,'$', 'M', 'k', 'NaN'

````python
playstore_df['Installs'].unique()
````

````python
#Clean special characters  from installs column


playstore_df['Installs'] = playstore_df['Installs'].apply(lambda x: x.replace(',',''))
playstore_df['Installs'] = playstore_df['Installs'].apply(lambda x: x.replace('+',''))
playstore_df['Installs'] = playstore_df['Installs'].astype(dtype = 'int')
````

````python
playstore_df.head()
````

````python
#cleaning special characters from size column 
playstore_df['Size'].unique()
````

````python
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: x.replace('M', '*1000'))
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: x.replace('k', ''))
playstore_df['Size'].replace('Varies with device', '-1', inplace = True) # changing the text Varies with device to -1 to perform eval operation
playstore_df['Size'] = playstore_df['Size'].apply(lambda x: eval(x))
````

````python
playstore_df['Size'] = playstore_df['Size'].replace(-1,np.nan) #Changing the values to null then we can fill them with mean value
playstore_df['Size'].fillna(np.mean(playstore_df['Size']), inplace = True)
````

````python
playstore_df.head()
````

````python
#removing $ sign from Price column
playstore_df[playstore_df['Price']!='0']
````

````python
playstore_df['Price'] = playstore_df['Price'].apply(lambda x: x.replace('$', ''))
playstore_df[playstore_df['Price']!='0']
````

````python
playstore_df['Price'] = playstore_df['Price'].astype(float)
playstore_df.head()
````

## Grouping basis on rating- Adding new column

````python
def Rat_app(val):
  ''''
  This function help to categories the rating from 1 to 5
  as Top_rated,Above_average,Average & below Average, worst
  '''
  if val>=4:
    return 'Top rated'
  elif val>=3 and val<4:
    return 'Above Average'
  elif val>=2 and val<3:
    return 'Average'
  else:
    return 'Below Average'
````

````python
# Applying grouped_rating function
playstore_df['Rating_group']=playstore_df['Rating'].apply(lambda x: Rat_app(x))
````
